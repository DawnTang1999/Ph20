```python
import numpy as np
from math import*
import matplotlib.pyplot as plt
```

```python
def curves(fx, fy, Ax, Ay, phi, dt, N):
    X = Ax * np.cos(np.arange(N)*dt*2*pi*fx)
    Y = Ay * np.sin(np.arange(N)*dt*2*pi*fy + phi)
    Z = X + Y
    np.savetxt('graphs_xyz.txt', (X,Y,Z), header='X is the first part, Y the second and Z the third')
    plt.plot(X,Y)
    plt.show()
    plt.plot(Z)
    plt.show()
```

```python
curves(2,3,1,1,0,0.01,200)
```

```python
curves(3,2,1,1,0,0.01,500)
```

```python
curves(2,2,1,1,0,0.01,1000)
```

```python
curves(2,10,1,1,0,0.001,2500)
```
